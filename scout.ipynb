# ***Main Players and Fixtures Data Read*** ---->

```python
# Configs
team = "Liverpool" # team to pull players data of
season = 2022 # season to pull players and teams stats of
league = 39 # league ID to pull from, current values: {39:premier league}
```

```python
# Fixtures for Events Data
## Main Entry point to get match summary and drill down with fixture ids
url = "https://v3.football.api-sports.io/fixtures?league={}&season={}".format(league,season)
fixtures_response = requests.get(url,headers=headers_api_sport)
fixtures_dat = pd.json_normalize(fixtures_response.json()['response'])
fixtures_dat.head()
```

```python
# Liverpools Fixtures for the 2022 season
liverpool_fixtures = fixtures_dat[(fixtures_dat['teams.home.name'] == team) | (fixtures_dat['teams.away.name'] == team)]
liverpool_fixtures_id = liverpool_fixtures[['fixture.id','teams.home.name','teams.away.name']]
liverpool_fixtures_id.head()
```

```python
print("Number of Liverpool fixtures to read: ",len(liverpool_fixtures_id))
```

```python
# Read player statistics for Liverpool
## There is a API call rate limit of 10 requests per minute
## Add sleep for a minute

player_stat_home = dict()
player_stat_away = dict()
#home Games
for fixture in liverpool_fixtures_id[liverpool_fixtures_id['teams.home.name'] == "Liverpool"]['fixture.id']:
    player_stat_url = "https://v3.football.api-sports.io/fixtures/players?fixture={}".format(fixture)
    player_stat_home[fixture] = requests.get(player_stat_url,headers=headers_api_sport)
    
    time.sleep(15)
    
#Away Games
for fixture in liverpool_fixtures_id[liverpool_fixtures_id['teams.home.name'] != "Liverpool"]['fixture.id']:
    player_stat_url = "https://v3.football.api-sports.io/fixtures/players?fixture={}".format(fixture)
    player_stat_away[fixture] = requests.get(player_stat_url,headers=headers_api_sport)
    time.sleep(15)
```

```python
# Example to get each players match statistic for a given team:
# Multiple levels to wrangle json data
player_stats_home = dict()
player_stats_away = dict()
#For Home Games, we can get the first 1st players statistics row and 2nd for away fixtures
for k,v in player_stat_home.items():
    print("Checking for Home team")
    try:
        print(pd.json_normalize(v.json()['response'])['players'])
        player_stats_home[k] = pd.concat([pd.json_normalize(pd.json_normalize(v.json()['response'])['players'][0])[['player.id','player.name']],pd.json_normalize(pd.json_normalize(pd.json_normalize(pd.json_normalize(v.json()['response'])['players'][0])['statistics']).rename(columns = {0:"player_stats"})['player_stats'])],axis = 1)
    except:
        print("problematic fixture id: ", k)
        continue
# Get 2nd for Away fixtures
for k,v in player_stat_away.items():
    print("Checking for Away team")
    try:
        print(pd.json_normalize(v.json()['response'])['players'])
        player_stats_away[k] = pd.concat([pd.json_normalize(pd.json_normalize(v.json()['response'])['players'][1])[['player.id','player.name']],pd.json_normalize(pd.json_normalize(pd.json_normalize(pd.json_normalize(v.json()['response'])['players'][1])['statistics']).rename(columns = {0:"player_stats"})['player_stats'])],axis = 1)
    except:
        print("problematic fixture id: ", k)
        continue
```

```python
for k,v in player_stats_home.items():
    v['fixture'] = k
for k,v in player_stats_away.items():
    v['fixture'] = k
```

```python
liverpool_player_stats = pd.concat(player_stats_home.values(),axis=0)
liverpool_player_stats = pd.concat([liverpool_player_stats] + list(player_stats_away.values()),axis = 0)
liverpool_player_stats.shape
```
